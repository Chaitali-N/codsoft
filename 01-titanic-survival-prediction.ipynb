# From "Ticket" to "is_group"

```python
ticket = tt.Ticket.value_counts(dropna=False)
ticket
```

```python
ticket[ticket!=1]
```

```python
group_list = list(ticket[ticket!=1].index)
group_list
```

```python
tt['is_group'] = tt.Ticket.isin(group_list)*1
tt['is_group']
```

```python
sns.countplot(data=tt, x='is_group',hue='Survived')
```

# From "SibSp" and "Parch" "to is_alone"

```python
tt.SibSp
```

```python
tt.Parch
```

```python
tt['is_alone'] =((tt.SibSp == 0) & (tt.Parch == 0))*1
tt.is_alone
```

```python
sns.countplot(data=tt, x='is_alone',hue='Survived')
```
